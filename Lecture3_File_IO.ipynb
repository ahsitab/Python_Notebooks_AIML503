## Part 1: Reading and Writing Text Files

### File Operations Overview
- **Modes:** `'r'` (read), `'w'` (write), `'a'` (append), `'r+'` (read+write)
- Always close files or use `with` statement
- Text files store human-readable characters

### 1. Creating a Sample Text File for Practice

```python
# First, let's create a sample text file to work with
sample_text = """Hello, World!
This is a sample text file.
Python makes file handling easy.
Line 4: Learning is fun!
Line 5: Practice makes perfect."""

# Write sample file
with open('sample.txt', 'w') as file:
    file.write(sample_text)

print("✓ Sample file 'sample.txt' created successfully!")
```

### 2. Reading Entire File at Once

```python
# Method 1: read() - reads entire file as a single string
print("=" * 50)
print("Method 1: read() - Entire file at once")
print("=" * 50)

with open('sample.txt', 'r') as file:
    content = file.read()
    print(content)
    print(f"\nTotal characters: {len(content)}")
```

```python
# Method 2: readlines() - returns list of lines
print("=" * 50)
print("Method 2: readlines() - List of lines")
print("=" * 50)

with open('sample.txt', 'r') as file:
    lines = file.readlines()
    print(f"Type: {type(lines)}")
    print(f"Number of lines: {len(lines)}")
    for i, line in enumerate(lines, 1):
        print(f"  Line {i}: {line.strip()}")
```

### **Reading entire file at once or reading list of lines is not memory efficient.**

```python
# Method 3: Iterating line by line (memory efficient for large files)
print("=" * 50)
print("Method 3: Line by line iteration")
print("=" * 50)

with open('sample.txt', 'r') as file:
    line_number = 0
    for line in file:
        line_number += 1
        # Process each line
        print(f"Processing line {line_number}: {line.strip()}")
```

```python
import csv

print("Reading CSV with csv.reader():")
print("=" * 50)

with open('students.csv', 'r') as file:
    csv_reader = csv.reader(file)

    # Read header
    header = next(csv_reader)
    print(f"Headers: {header}")
    print("-" * 50)

    # Read data rows
    for row in csv_reader:
        print(f"{row[0]:10} | Age: {row[1]:2} | City: {row[2]:15} | Score: {row[3]:3} | Grade: {row[4]}")
```

### 7. Reading CSV as Dictionaries with csv.DictReader()

```python
print("Reading CSV with csv.DictReader():")
print("=" * 50)

students_data = []

with open('students.csv', 'r') as file:
    csv_reader = csv.DictReader(file)

    for row in csv_reader:
        # Convert Score to integer for calculations
        row['Score'] = int(row['Score'])
        row['Age'] = int(row['Age'])
        students_data.append(row)

# Now we can work with the data easily
print("Students data loaded as dictionaries:")
for student in students_data[:3]:  # First 3 students
    print(f"  {student['Name']}: Score={student['Score']}, Grade={student['Grade']}")

# Calculate statistics
avg_score = sum(s['Score'] for s in students_data) / len(students_data)
avg_age = sum(s['Age'] for s in students_data) / len(students_data)
print(f"\n📊 Statistics:")
print(f"  Total Students: {len(students_data)}")
print(f"  Average Score: {avg_score:.1f}")
print(f"  Average Age: {avg_age:.1f}")
```

- Lightweight data interchange format
- Easy for humans to read and write
- Easy for machines to parse and generate
- Language-independent

| JSON Type | Python Type |
|-----------|-------------|
| object | dict |
| array | list |
| string | str |
| number (int) | int |
| number (real) | float |
| true | True |
| false | False |

```python
# Working with JSON in Python
import json

# Python dictionary
person = {
    "name": "Alice Johnson",
    "age": 28,
    "city": "San Francisco",
    "is_student": False,
    "hobbies": ["reading", "hiking", "photography"],
    "education": {
        "degree": "Bachelor's",
        "major": "Computer Science",
        "year": 2020
    },
    "skills": None
}

# Convert Python dict to JSON string
json_string = json.dumps(person, indent=2)
print("Python dict → JSON string:")
print(json_string)
print(f"\nType: {type(json_string)}")
```

## Part 4: Exception Handling with try-except

### Why Exception Handling?
- Programs should handle errors gracefully
- Prevents crashes from unexpected input
- Provides meaningful error messages to users
- Essential for robust applications

### Common Python Exceptions

| Exception | Description |
|-----------|-------------|
| `TypeError` | Wrong type operation |
| `ValueError` | Right type, wrong value |
| `FileNotFoundError` | File doesn't exist |
| `KeyError` | Dictionary key not found |
| `IndexError` | List index out of range |
| `ZeroDivisionError` | Division by zero |
| `AttributeError` | Object has no such attribute |
| `ImportError` | Module not found |

```python
# Basic try-except structure
print("Basic try-except examples:")
print("=" * 50)

# Example 1: Handling division by zero
try:
    result = 10 / 0
except ZeroDivisionError:
    print("❌ Error: Cannot divide by zero!")

# Example 2: Handling file not found
try:
    with open('nonexistent_file.txt', 'r') as f:
        content = f.read()
except FileNotFoundError:
    print("❌ Error: File not found!")

# Example 3: Handling type errors
try:
    result = "hello" + 5
except TypeError as e:
    print(f"❌ TypeError: {e}")

print("✓ Program continues execution after handling errors!")
```

```python
# Handling multiple exceptions
def safe_convert_to_int(value):
    """Safely convert a value to integer."""
    try:
        return int(value)
    except ValueError:
        print(f"  ❌ '{value}' cannot be converted to integer")
        return None
    except TypeError:
        print(f"  ❌ Invalid type for conversion: {type(value)}")
        return None

# Test various inputs
test_values = ["123", "45.6", "abc", True, None, [1, 2, 3]]

print("Testing value conversion:")
for val in test_values:
    result = safe_convert_to_int(val)
    if result is not None:
        print(f"  ✓ '{val}' → {result}")
    print()
```

```python
# try-except-else-finally complete structure
print("Complete try-except-else-finally structure:")
print("=" * 50)

def read_number_from_file(filename):
    """
    Try to read a number from a file.
    Demonstrates try-except-else-finally.
    """
    try:
        # Attempt to open and read file
        with open(filename, 'r') as f:
            content = f.read().strip()
        number = int(content)
    except FileNotFoundError:
        print(f"  ❌ File '{filename}' not found")
        return None
    except ValueError:
        print(f"  ❌ File content is not a valid number")
        return None
    else:
        # Executes only if no exception occurred
        print(f"  ✅ Successfully read: {number}")
        return number
    finally:
        # Always executes - cleanup code
        print(f"  🔄 File operation completed for '{filename}'")

# Test with different scenarios
print("\nTest 1: Valid file")
with open('number.txt', 'w') as f:
    f.write("42")
read_number_from_file('number.txt')

print("\nTest 2: Non-existent file")
read_number_from_file('no_file.txt')

print("\nTest 3: Invalid content")
with open('bad_number.txt', 'w') as f:
    f.write("not_a_number")
read_number_from_file('bad_number.txt')
```

```python
# Raising exceptions manually
def validate_age(age):
    """Validate age input."""
    if not isinstance(age, (int, float)):
        raise TypeError(f"Age must be a number, got {type(age).__name__}")
    if age < 0:
        raise ValueError("Age cannot be negative")
    if age > 150:
        raise ValueError(f"Age {age} is unrealistic")
    return True

def process_user(age):
    """Process user age with proper error handling."""
    try:
        validate_age(age)
        category = "child" if age < 13 else "teen" if age < 20 else "adult" if age < 65 else "senior"
        return f"User is a {category} (age {age})"
    except TypeError as e:
        return f"Type Error: {e}"
    except ValueError as e:
        return f"Value Error: {e}"

# Test validation
test_ages = [25, -5, 200, "thirty", 45.5, None]
print("Age Validation Tests:")
print("=" * 50)
for age in test_ages:
    result = process_user(age)
    print(f"  Input: {str(age):10} → {result}")
```

```python
# Real-world example: Safe file reader
def safe_read_csv(filename):
    """
    Safely read a CSV file with comprehensive error handling.
    """
    try:
        with open(filename, 'r') as file:
            reader = csv.DictReader(file)
            data = list(reader)

            if not data:
                print(f"⚠️  Warning: File '{filename}' is empty")
                return []

            print(f"✅ Successfully loaded {len(data)} rows from '{filename}'")
            return data

    except FileNotFoundError:
        print(f"❌ Error: File '{filename}' does not exist")
        return []
    except PermissionError:
        print(f"❌ Error: Permission denied for '{filename}'")
        return []
    except csv.Error as e:
        print(f"❌ Error: Invalid CSV format in '{filename}': {e}")
        return []
    except Exception as e:
        print(f"❌ Unexpected error reading '{filename}': {e}")
        return []

# Test the safe reader
print("Testing safe CSV reader:")
print("=" * 50)

print("\n1. Reading valid file:")
data1 = safe_read_csv('students.csv')
print(f"   Returned {len(data1)} records")

print("\n2. Reading non-existent file:")
data2 = safe_read_csv('ghost.csv')

print("\n3. Reading empty file:")
with open('empty.csv', 'w') as f:
    pass
data3 = safe_read_csv('empty.csv')
```

### Creating Custom Exceptions

```python
# Custom exception classes
class InvalidDataError(Exception):
    """Custom exception for invalid data."""
    pass

class DataValidationError(Exception):
    """Custom exception for data validation failures."""
    def __init__(self, message, field=None, value=None):
        self.field = field
        self.value = value
        super().__init__(message)

def validate_student_record(record):
    """Validate a student record."""
    required_fields = ['Name', 'Age', 'Score']

    for field in required_fields:
        if field not in record:
            raise DataValidationError(
                f"Missing required field: {field}",
                field=field
            )

    try:
        age = int(record['Age'])
        if age < 0 or age > 120:
            raise DataValidationError(
                f"Invalid age: {age}",
                field='Age',
                value=age
            )
    except ValueError:
        raise DataValidationError(
            f"Age must be a number: {record['Age']}",
            field='Age',
            value=record['Age']
        )

    return True

# Test validation
test_records = [
    {'Name': 'Alice', 'Age': '25', 'Score': '85'},
    {'Name': 'Bob', 'Score': '92'},  # Missing Age
    {'Name': 'Charlie', 'Age': '-5', 'Score': '78'},  # Invalid age
    {'Name': 'Diana', 'Age': 'abc', 'Score': '95'},  # Non-numeric age
]

print("Student Record Validation:")
print("=" * 50)
for i, record in enumerate(test_records, 1):
    print(f"\nRecord {i}: {record.get('Name', 'Unknown')}")
    try:
        validate_student_record(record)
        print(f"  ✅ Valid record")
    except DataValidationError as e:
        print(f"  ❌ Validation Error: {e}")
        if e.field:
            print(f"     Problem field: {e.field}")
        if e.value is not None:
            print(f"     Problem value: {e.value}")
```

## Part 5: Debugging Basics

### Debugging Techniques
- Print statements for tracing
- Using assertions
- Python's `pdb` debugger
- Logging module

```python
# Technique 1: Strategic print statements
print("Debugging Technique 1: Print Statements")
print("=" * 50)

def calculate_average_with_debug(numbers):
    """Calculate average with debug prints."""
    print(f"[DEBUG] Input: {numbers}")
    print(f"[DEBUG] Type: {type(numbers)}")

    if not numbers:
        print("[DEBUG] Empty list detected")
        return 0

    total = sum(numbers)
    count = len(numbers)
    print(f"[DEBUG] Total: {total}, Count: {count}")

    average = total / count
    print(f"[DEBUG] Average: {average}")

    return average

# Test
result = calculate_average_with_debug([85, 92, 78, 90])
print(f"Result: {result}")
```

```python
# Technique 2: Using assert statements
print("Debugging Technique 2: Assertions")
print("=" * 50)

def process_scores(scores):
    """Process student scores with assertions."""
    # Assertions check conditions that should ALWAYS be true
    assert isinstance(scores, list), "Scores must be a list"
    assert len(scores) > 0, "Scores list cannot be empty"
    assert all(isinstance(s, (int, float)) for s in scores), "All scores must be numbers"
    assert all(0 <= s <= 100 for s in scores), "Scores must be between 0 and 100"

    avg = sum(scores) / len(scores)
    return {
        'average': avg,
        'max': max(scores),
        'min': min(scores),
        'count': len(scores)
    }

# Test with valid data
try:
    result = process_scores([85, 92, 78, 90])
    print(f"✅ Valid: {result}")
except AssertionError as e:
    print(f"❌ Assertion failed: {e}")

# Test with invalid data
print("\nTesting invalid data:")
invalid_cases = [
    "not a list",
    [],
    [85, "92", 78],
    [85, 150, 78]
]

for case in invalid_cases:
    try:
        process_scores(case)
        print(f"  Input {case}: ✅ Passed")
    except AssertionError as e:
        print(f"  Input {case}: ❌ {e}")
```

```python
# Technique 3: Using logging module
import logging

# Configure logging
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s - %(levelname)s - %(message)s',
    datefmt='%H:%M:%S'
)

print("Debugging Technique 3: Logging")
print("=" * 50)

def analyze_data(data):
    """Analyze data with proper logging."""
    logging.info(f"Starting data analysis")
    logging.debug(f"Input data: {data}")

    if not data:
        logging.warning("Empty dataset provided")
        return None

    try:
        # Remove outliers (values > 3 standard deviations)
        import statistics
        mean = statistics.mean(data)
        stdev = statistics.stdev(data)
        logging.debug(f"Mean: {mean:.2f}, StDev: {stdev:.2f}")

        threshold = 3 * stdev
        filtered = [x for x in data if abs(x - mean) <= threshold]

        removed = len(data) - len(filtered)
        if removed > 0:
            logging.warning(f"Removed {removed} outliers")

        logging.info(f"Analysis complete. {len(filtered)} records remain.")
        return filtered

    except Exception as e:
        logging.error(f"Analysis failed: {e}")
        return None

# Test with normal data
print("\nTest 1: Normal data")
normal_data = [1, 2, 3, 4, 5, 3, 4, 2, 3, 4]
result = analyze_data(normal_data)
print(f"  Result: {result}")

# Test with outliers
print("\nTest 2: Data with outliers")
data_with_outliers = [1, 2, 3, 4, 5, 3, 4, 2, 3, 4, 100]  # 100 is an outlier
result = analyze_data(data_with_outliers)
print(f"  Result: {result}")

# Test with empty data
print("\nTest 3: Empty data")
result = analyze_data([])
print(f"  Result: {result}")
```

## Part 6: Importance of Clean Datasets in AI

### Why Data Quality Matters
- **Garbage In, Garbage Out (GIGO):** Poor data leads to poor models
- AI models learn patterns from data
- Dirty data introduces bias and errors
- Data cleaning is 60-80% of data science work

### Common Data Quality Issues

1. **Missing Values:** Empty or NaN entries
2. **Duplicate Records:** Same data entered multiple times
3. **Inconsistent Formats:** Different date formats, capitalization
4. **Outliers:** Extreme values that skew results
5. **Incorrect Data Types:** Numbers stored as strings
6. **Invalid Data:** Values outside expected ranges

```python
# Data Quality Analysis Function
def analyze_data_quality(data):
    """
    Analyze dataset for common quality issues.
    Returns a quality report.
    """
    total_rows = len(data)
    quality_report = {
        'total_rows': total_rows,
        'issues_found': {},
        'quality_score': 100
    }

    # Check for missing values
    for row in data:
        for key, value in row.items():
            if value is None or value == '' or value == ' ':
                if 'missing_values' not in quality_report['issues_found']:
                    quality_report['issues_found']['missing_values'] = []
                quality_report['issues_found']['missing_values'].append(
                    f"{key} is empty for {row.get('Name', 'Unknown')}"
                )

    # Check for duplicates (based on Name and Email)
    seen = {}
    for i, row in enumerate(data):
        key = (row.get('Name', ''), row.get('Email', ''))
        if key in seen:
            if 'duplicates' not in quality_report['issues_found']:
                quality_report['issues_found']['duplicates'] = []
            quality_report['issues_found']['duplicates'].append(
                f"Duplicate: {key[0]} ({key[1]}) at row {i}"
            )
        seen[key] = i

    # Check for invalid age
    for row in data:
        age = row.get('Age', '')
        if age:
            try:
                age_num = int(age)
                if age_num < 0 or age_num > 120:
                    if 'invalid_age' not in quality_report['issues_found']:
                        quality_report['issues_found']['invalid_age'] = []
                    quality_report['issues_found']['invalid_age'].append(
                        f"Invalid age {age} for {row.get('Name', 'Unknown')}"
                    )
            except ValueError:
                if 'invalid_age' not in quality_report['issues_found']:
                    quality_report['issues_found']['invalid_age'] = []
                quality_report['issues_found']['invalid_age'].append(
                    f"Non-numeric age '{age}' for {row.get('Name', 'Unknown')}"
                )

    # Check email format
    for row in data:
        email = row.get('Email', '')
        if email and '@' not in email:
            if 'invalid_email' not in quality_report['issues_found']:
                quality_report['issues_found']['invalid_email'] = []
            quality_report['issues_found']['invalid_email'].append(
                f"Invalid email '{email}' for {row.get('Name', 'Unknown')}"
            )

    # Calculate quality score
    total_issues = sum(len(v) for v in quality_report['issues_found'].values())
    quality_report['quality_score'] = max(0, 100 - (total_issues * 5))
    quality_report['total_issues'] = total_issues

    return quality_report

# Load dirty data
dirty_records = []
with open('dirty_data.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        dirty_records.append(row)

# Analyze quality
quality_report = analyze_data_quality(dirty_records)

print("📊 DATA QUALITY REPORT")
print("=" * 70)
print(f"Total Records: {quality_report['total_rows']}")
print(f"Quality Score: {quality_report['quality_score']}/100")
print(f"Issues Found: {quality_report['total_issues']}")

if quality_report['issues_found']:
    print("\n🔍 Detailed Issues:")
    for issue_type, issues in quality_report['issues_found'].items():
        print(f"\n  📍 {issue_type.replace('_', ' ').title()}: {len(issues)} issues")
        for issue in issues[:3]:  # Show first 3 of each type
            print(f"     • {issue}")
        if len(issues) > 3:
            print(f"     ... and {len(issues) - 3} more")
```

```python
# Data Cleaning Pipeline
print("🧹 DATA CLEANING PIPELINE")
print("=" * 70)

def clean_dataset(data):
    """
    Clean the dataset by fixing common issues.
    Returns cleaned data and cleaning log.
    """
    cleaning_log = []
    cleaned_data = []

    seen_records = set()

    for i, row in enumerate(data):
        modified = False
        original = dict(row)

        # 1. Standardize Name (capitalize)
        if row.get('Name'):
            new_name = row['Name'].strip().title()
            if new_name != row['Name']:
                cleaning_log.append(f"Row {i}: Standardized name '{row['Name']}' → '{new_name}'")
                row['Name'] = new_name
                modified = True

        # 2. Handle missing Age (use median or skip)
        if not row.get('Age') or row['Age'] == '':
            cleaning_log.append(f"Row {i}: Missing age for {row['Name']} - will skip")
            continue  # Skip rows with missing critical data

        # 3. Validate and fix Age
        try:
            age = int(row['Age'])
            if age < 0 or age > 120:
                cleaning_log.append(f"Row {i}: Invalid age {age} for {row['Name']} - will skip")
                continue
            row['Age'] = str(age)
        except ValueError:
            cleaning_log.append(f"Row {i}: Non-numeric age for {row['Name']} - will skip")
            continue

        # 4. Validate Email
        if row.get('Email') and '@' not in row['Email']:
            cleaning_log.append(f"Row {i}: Invalid email for {row['Name']} - will skip")
            continue

        # 5. Handle missing Salary
        if not row.get('Salary') or row['Salary'] == '':
            cleaning_log.append(f"Row {i}: Missing salary for {row['Name']}")
            row['Salary'] = '0'  # Default value
            modified = True

        # 6. Validate Salary is numeric
        try:
            float(row['Salary'])
        except ValueError:
            cleaning_log.append(f"Row {i}: Invalid salary for {row['Name']} - will skip")
            continue

        # 7. Check for duplicates (Name + Email)
        record_key = (row.get('Name', ''), row.get('Email', ''))
        if record_key in seen_records:
            cleaning_log.append(f"Row {i}: Duplicate record for {row['Name']} - removed")
            continue
        seen_records.add(record_key)

        # 8. Standardize Date format (handle different separators)
        if row.get('Join_Date'):
            row['Join_Date'] = row['Join_Date'].replace('/', '-')

        if modified:
            cleaning_log.append(f"Row {i}: Modified record for {row['Name']}")

        cleaned_data.append(row)

    return cleaned_data, cleaning_log

# Clean the dataset
cleaned_data, cleaning_log = clean_dataset(dirty_records)

print(f"\nOriginal records: {len(dirty_records)}")
print(f"Cleaned records: {len(cleaned_data)}")
print(f"Records removed: {len(dirty_records) - len(cleaned_data)}")

print(f"\n📝 Cleaning Log ({len(cleaning_log)} operations):")
print("-" * 70)
for log_entry in cleaning_log[:10]:  # Show first 10 entries
    print(f"  {log_entry}")
if len(cleaning_log) > 10:
    print(f"  ... and {len(cleaning_log) - 10} more operations")

# Save cleaned data
if cleaned_data:
    with open('cleaned_data.csv', 'w', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=cleaned_data[0].keys())
        writer.writeheader()
        writer.writerows(cleaned_data)
    print(f"\n✅ Cleaned data saved to 'cleaned_data.csv'")
```

## 📚 Comprehensive Exercise

### Task: Build a Complete Data Processing Pipeline

Create a system that:
1. Reads a CSV file with potential errors
2. Validates and cleans the data
3. Handles all exceptions gracefully
4. Saves the cleaned data as JSON
5. Generates a quality report

```python
# COMPREHENSIVE EXERCISE SOLUTION
import json
import csv
from datetime import datetime

class DataPipeline:
    """Complete data processing pipeline."""

    def __init__(self, input_file):
        self.input_file = input_file
        self.raw_data = []
        self.cleaned_data = []
        self.quality_report = {}
        self.errors = []

    def load_data(self):
        """Load data from CSV file."""
        try:
            with open(self.input_file, 'r') as f:
                reader = csv.DictReader(f)
                self.raw_data = list(reader)
            print(f"✅ Loaded {len(self.raw_data)} records from {self.input_file}")
            return True
        except FileNotFoundError:
            self.errors.append(f"File not found: {self.input_file}")
            print(f"❌ File not found: {self.input_file}")
            return False
        except Exception as e:
            self.errors.append(str(e))
            print(f"❌ Error loading file: {e}")
            return False

    def clean_data(self):
        """Clean and validate the loaded data."""
        if not self.raw_data:
            print("⚠️  No data to clean")
            return

        total = len(self.raw_data)
        removed = 0
        fixed = 0

        for row in self.raw_data:
            try:
                # Validate required fields
                required = ['Name', 'Age', 'Email']
                for field in required:
                    if not row.get(field):
                        raise ValueError(f"Missing required field: {field}")

                # Clean and validate age
                age = int(row['Age'])
                assert 0 <= age <= 120, f"Invalid age: {age}"

                # Validate email
                email = row.get('Email', '')
                assert '@' in email, f"Invalid email: {email}"

                self.cleaned_data.append(row)

            except (ValueError, AssertionError) as e:
                removed += 1
                self.errors.append(f"Row removed: {e}")

        print(f"✅ Cleaning complete: {len(self.cleaned_data)} records retained, {removed} removed")

    def generate_report(self):
        """Generate quality report."""
        self.quality_report = {
            'timestamp': datetime.now().isoformat(),
            'input_file': self.input_file,
            'total_records': len(self.raw_data),
            'cleaned_records': len(self.cleaned_data),
            'removed_records': len(self.raw_data) - len(self.cleaned_data),
            'quality_score': round(len(self.cleaned_data) / max(len(self.raw_data), 1) * 100, 1),
            'errors': self.errors[:5]  # First 5 errors
        }
        print(f"✅ Report generated")

    def save_output(self, output_file):
        """Save cleaned data and report as JSON."""
        output = {
            'quality_report': self.quality_report,
            'cleaned_data': self.cleaned_data
        }
        try:
            with open(output_file, 'w') as f:
                json.dump(output, f, indent=2)
            print(f"✅ Output saved to {output_file}")
        except Exception as e:
            print(f"❌ Error saving output: {e}")

    def run(self, output_file='pipeline_output.json'):
        """Run the complete pipeline."""
        print("\n🚀 Starting Data Pipeline")
        print("=" * 50)

        if self.load_data():
            self.clean_data()
            self.generate_report()
            self.save_output(output_file)

        print("\n📊 Final Report:")
        print(json.dumps(self.quality_report, indent=2))
        print("\n✅ Pipeline complete!")

# Run the pipeline
pipeline = DataPipeline('dirty_data.csv')
pipeline.run('cleaned_output.json')

# Verify the output
print("\n📁 Verifying output file:")
with open('cleaned_output.json', 'r') as f:
    output = json.load(f)
    print(f"  Records in output: {len(output['cleaned_data'])}")
    print(f"  Quality score: {output['quality_report']['quality_score']}%")
```
